****Data have bad destribution which have 280000 in 0 label and have less than 500 in 1 so to change the destribution we will take only 10000 from 0**** 

```python
class_0=df[df['Class']==0]
class_0=class_0.head(10000)
class_1=df[df["Class"]==1]
df_new=pd.concat([class_0,class_1],axis=0)
```

```python
df_new['Class'].value_counts()
```

```python
x_u=df_new.drop('Class',axis=1)
y_u=df_new['Class']
```

```python
#unique selection
from sklearn.model_selection import train_test_split
x_train_u,x_test_u,y_train_u,y_test_u=train_test_split(x_u,y_u,test_size=.2,random_state=42,stratify=y_u)
print("x_train_u: ",x_train_u.shape)
print("y_train_u: ",y_train_u.shape)
print("x_test_u: ",x_test_u.shape)
print("y_test_u: ",y_test_u.shape)
```

```python
#work in all data
x_a=df.drop("Class",axis=1)
y_a=df['Class']
x_train_a,x_test_a,y_train_a,y_test_a=train_test_split(x_a,y_a,test_size=.2,random_state=42)
print("x_train_a: ",x_train_a.shape)
print("y_train_a: ",y_train_a.shape)
print("x_test_a: ",x_test_a.shape)
print("y_test_a: ",y_test_a.shape)
```

```python
#work in SMOTE
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_train_sm,y_train_sm=smote.fit_resample(x_train_u,y_train_u)
```

```python
from sklearn.metrics import classification_report
def train_model(model,x_train,y_train,x_test,y_test):
    model=model
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    report=classification_report(y_test,pred)
    print(report)
    print("-"*50)
```

```python
from sklearn.svm import SVC
#svc and unique data
print("svc and unique data")
train_model(SVC(),x_train_u,y_train_u,x_test_u,y_test_u)
#svc and all data
print("svc and all data")
train_model(SVC(),x_train_a,y_train_a,x_test_a,y_test_a)
#svc and SMOTE data
print("svc and SMOTE data")
train_model(SVC(),x_train_sm,y_train_sm,x_test_a,y_test_a)
```

```python
from sklearn.ensemble import RandomForestClassifier
#Randomforest and unique data
print("Randomforest and unique data")
train_model(RandomForestClassifier(),x_train_u,y_train_u,x_test_u,y_test_u)
#Randomforest and all data
print("Randomforest and all data")
train_model(RandomForestClassifier(),x_train_a,y_train_a,x_test_a,y_test_a)
#Randomforest and SMOTE data
print("Randomforest and SMOTE data")
train_model(RandomForestClassifier(),x_train_sm,y_train_sm,x_test_a,y_test_a)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
#Gradient Boosting and unique data
print("Gradient Boosting and unique data")
train_model(GradientBoostingClassifier(),x_train_u,y_train_u,x_test_u,y_test_u)
#Gradient Boosting and all data
print("Gradient Boosting and all data")
train_model(GradientBoostingClassifier(),x_train_a,y_train_a,x_test_a,y_test_a)
#Gradient Boosting and SMOTE data
print("Gradient Boosting and SMOTE data")
train_model(GradientBoostingClassifier(),x_train_sm,y_train_sm,x_test_a,y_test_a)
```

```python
from sklearn.metrics import accuracy_score
model=GradientBoostingClassifier()
model.fit(x_train_u,y_train_u)
pred=model.predict(x_test_u)
accuracy_score(y_test_u,pred)
```

```python
import joblib
joblib.dump(model,"model.pkl")
```
